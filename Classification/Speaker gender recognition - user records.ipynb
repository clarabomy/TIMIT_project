# Speaker gender recognition - user records

In this notebook, we will test the model obtained with the neural network for the classification by gender on our own voices.

# Importing useful libraries

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from scipy.io import wavfile as wav
import IPython.display as ipd
import random, pyaudio, wave, utils, os
import seaborn as sns
import utils  #utils contains all the functions we have created or inspired from
```

We import the model trained with the TIMIT dataset.

```python
from keras.models import Model, load_model
trained_model = "gender_classifier.h5"
model = load_model(trained_model)
```

We create a folder that will contain all the user records. To keep track, each created audio file is saved in the audio_test folder.

```python
directory = "audio_test/"
if not os.path.exists(directory):
    os.makedirs(directory)
```

### Creation of the recording audio file.

We used the record function of the utils.py file to create WAV files with the computer's microphone.

```python
filename = "test.wav"
path = os.path.join(directory, filename)
utils.record(path)
ipd.Audio(path)
```

### Predictions of the model

```python
X_test = utils.extract_features(path)
X_test = X_test.reshape((1, 40))
```

```python
prediction = model.predict(X_test)
labels = ["male","female"]
n_test = X_test.shape[0]
print("filename :", path)
print("model prediction :", labels[(int(round(prediction[0][0])))], "-",prediction[0][0])
ipd.Audio(path)
```

Thus, we noticed that our model recognizes more easily the voices of men. Regarding the recognition of women's voices, the result is rather variable and will depend on the intonation and the tone used. This surprised us because we had made sure that the model was learned from as many men's and women's voice recordings.
The model is more easily able to detect the speaker's gender when the language used is English: this result is hardly surprising because all audio files in the TIMIT dataset are in English.
