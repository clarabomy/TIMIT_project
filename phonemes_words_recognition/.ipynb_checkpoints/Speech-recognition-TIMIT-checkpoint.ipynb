# STEP 1: Acoustic Features for Speech Recognition

For this project, we won't use the audio waveform as input to our models. Instead, we provide code that first performs a pre-processing step to convert the audio file to a feature representation which is commonly used for speech recognition models. Thus, our models will accept the feature representation as input.

### Mel-Frequency Cepstral Coefficients (MFCCs)

We will normalize all the MFCCs of the audio files.

The main idea behind MFCC features is that at each time window, the MFCC feature yields a feature vector that characterizes the sound within the window. Note that the MFCC feature is quite lower-dimensional which will help an  model to avoid overfitting to the training dataset.

```python
from data_generator import plot_mfcc_feature

# plot normalized MFCC
plot_mfcc_feature(vis_mfcc_feature_wrd)
# print shape of MFCC
display(Markdown('**Shape of MFCC** : ' + str(vis_mfcc_feature_wrd.shape)))
```

# STEP 2: Neural Networks for Acoustic Modeling

In this section, you will experiment with various neural network architectures. All models will be specified in the sample_models.py file. 

### Model 1: RNN 

Given their effectiveness in modeling sequential data, we will use recurrent neural networks. The models will take the time sequence of audio features as input.

At each time step, the speaker pronounces one of 28 possible characters, including each of the 26 letters in the English alphabet, along with a space character (" "), and an apostrophe ('). As shown bellow, the output of the RNN at each time step is a vector of probabilities with 29 entries, where the  i -th entry encodes the probability that the  i -th character is spoken in the time sequence. 


<img src="images/rnn_model.png" width="60%">


The next figure shows an equivalent, rolled depiction of the RNN that shows the TimeDistrbuted dense and output layers in greater detail.  



<img src="images/rnn_model_unrolled.png" width="60%">


- The first layer of the neural network should be an RNN (SimpleRNN, LSTM, or GRU) that takes the time sequence of audio features as input. The BatchNormalization layer allows us to normalize the activations of the previous layer at each batch - the layer will apply a transformation that maintains the mean activation close to 0 and the activation standard deviation close to 1.


- We will use the output of our RNN as a hidden layer and the TimeDistributed layer to apply a Dense layer to each of the time steps in the RNN output. 


```python
rnn_model, phn = rnn_model(input_dim=13, 
                    units=200,
                    activation='relu',
                    phn=False)
```

Now, we train our model using the function train_model of the train_utils.py file. After the model has finised training, the loss history is saved in `model_rnn.pickle` and the model in `model_rnn.h5` in the `results` folder. 


```python
from datetime import datetime as dt
print ("Train STARTED at :", str(dt.now())[0:19])

if (phn):
    print ("\nYou're gonna work on phonemes.\n")
else:
    print ("\nYou're gonna work on words.\n")

train_model(input_to_softmax=rnn_model, phn=phn, 
            pickle_path='final\\rnn_wrd_10.pickle', 
            save_model_path='final\\rnn_wrd_10.h5',
            epochs= 10)

print ("Train FINISHED at :", str(dt.now())[0:19])
```
